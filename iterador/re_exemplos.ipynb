```python
import re   # se voce usar o r"" significa para ele read o texto entao sempre que quiser analisalo use r""
            # \d+ e para checar se tem um numero com 1 ou mais digitos no que voce quiser,
            # ex (r"\d+", formulario_seila_oq) ele vai checar os numero de 1 ou mais digitos do formulario_seila_oque
```

```python
texto = "Skibidi Gyat 676767676767"
resultado = re.search(r"\d+", texto)  # o serach vai procurar a primeria ocorrencia de algo que vc define,nesse caso usei \d+ ou seja, a primeira vez que aparece um numero

if resultado:
    print(resultado)   # dada
    
```

```python
texto1 = "Tenho 14 pessoas no meu porao, mas na minha outra casa tenho apenas 3"
resultado1 = re.findall(r"\d+", texto1)
if resultado1:
    print(resultado1)
```
