enumerate
- associa o valor e o indice de  uma lista


```python
frutas = ["banana", "laranja", "uva"]
```

```python

for indice, fruta in enumerate(frutas):
    print(f"{indice} - {fruta}")
```

```python
numeros = [1, 2, 6, 8, 3, 30]
```

```python
indice_valores = {indice: valor for indice, valor in enumerate(numeros)}
indice_valores
```

```python
indice_pares = {indice: valor for indice, valor in indice_valores.items() if indice % 2 == 0}
indice_pares
```
