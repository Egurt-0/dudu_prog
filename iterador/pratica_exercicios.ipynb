```python
def triplo(x):
    return x*3

def seila(funcao, lista):
    return [funcao(x) for x in lista]
```

```python
lista_numeros = [1, 2, 3, 4, 5]
triplo_dos_numeros = seila(triplo, lista_numeros)
triplo_dos_numeros
```

```python
palavras = [
    {"Nome": "piano" }, {"Nome": "guerra" },{"Nome": "avi√£o"}, {"Nome": "beleza"}, {"Nome": "sigilo"}]
```

```python
def comprimento_palavras(lista): # Dica, nao passar list como argumento, pois existe uma funcao que ja se chama list, a que transforma as coisas em uma lista, vc ja sabe
    return list(map(lambda u: len(u["Nome"]), lista)) 

print(comprimento_palavras(palavras))
```

```python
Lista_celsius = [0, 10, 20, 30, 40]
```

```python
def converter_fahrenheit(lista):
    return list(map(lambda u: u * 1.8 + 32, lista))

lista_temperaturas_convertidas = converter_fahrenheit(Lista_celsius)
print(lista_temperaturas_convertidas)
```
