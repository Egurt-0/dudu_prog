```python

class Contador:
    def __init__(self, limit):  
        self.limit = limit
        self.atual = 1


    def __iter__(self):
        return self
    
    def __next__(self):
        if self.atual > self.limit:
            raise StopIteration
        else:
            valor = self.atual
            self.atual += 1
            return valor
```

```python

                        # O numero que voce coloca dentro do parenteses define o objeto limit, ou seja se olhar o codigo vc ve que o objeto limit limita a iteracao("UUAUU que novidade")
contador = Contador(9)  # nao vai funcionar se voce tentar rodar apenas esse aq, pois ele nao vai achar a class Contador, 
                        #pois voce nao rodou ela, entao voce precisa rodar todos ao mesmo tempo, ISSO SERVER PARA TODOS SCRIPTS DESSE MODELO
for numero in contador:
    print(numero)
```
