```python
import itertools
```

```python
# Exemplo 1 - range()

for i in range(5):
    print(i)
```

```python
# ITERADOR DE LISTAS
frutas =["Banana", "Uva", "Limao", "Bezerro", "Curupira"]
for fruta in frutas:
    print(fruta)
```

```python
# Iterador de DICIONARIO
pessoa = {'nome': 'Alice', 'idade': 30}
for chave in pessoa:
    print(chave, pessoa[chave])
```

```python
# ENUMERATE()
for indice, fruta in enumerate(frutas):
    print(indice, fruta)
```

```python
# ZIP
nomes = ['robson', 'Alice', 'joao']
idades = [30, 25, 35]
for nome, idade in zip(nomes, idades):
    print(nome, idade)
```

```python
# Itertools
for i in itertools.count(start=10, step=2):
    if i >20:
        break
    print(i)
```
