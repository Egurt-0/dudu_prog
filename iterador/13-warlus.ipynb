Warlus
- esse operador é usado para atribuições de variaveis dentro de expressões
- permite que voce atribuia um valor a varaivel enquanto avalia a expressão

```python
numeros = [5, 23, 12, 63, 87, 67, 96, 73, 235, 73]
```

```python
# Filtrar e armazenar numeros maiores que 65

numeros_maiores_65 = [numero for numero in numeros if(maior := numero) > 65]
numeros_maiores_65
```
