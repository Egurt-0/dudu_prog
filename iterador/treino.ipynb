```python
class ContadorNumeros:
    def __init__(self, limite):     # lembre-se sempre de colocar as funcoes(DEFs) dentro da class, 
                                    #se nao elas nao vao servir de nada quando for chamar a classe ela nao vai conter as funcoes
        self.limite = limite
        self.atual = 1


    def __iter__(self):
        return self


    def __next__(self):
        if self.atual > self.limite:
            raise StopIteration
        else:
            valor = self.atual
            self.atual += 1
            return valor

        
```

```python
contador = ContadorNumeros(51)

for numero in contador:
    print(numero)
```
