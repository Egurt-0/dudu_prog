Map
 - funcao que aplica uma funcao a todos os itens de um iteravel (por exemplo uma lisa)
 - e uma maneira eficiente e consisa de transformar dados
 - minha explicacao  ------ entao o map ele vai ser utilizado para substituir o loop, ex: result =  map(quadrado, numeros) ele vai executar a funcao "quadrado" para cada item da lista numeros                                                               mas no final eu preciso definir a variavel result como algo, ex: list(result) estou definindo ela como uma lista
 tudo isso sem precisar criar im for numero in numeros:
                                        print(quadrado(numero))

```python
def quadrado(x):
    return x ** 2
     
```

```python
numeros = [1, 2, 3, 4, 5]
```

```python
for n in numeros:
    print(f"AQUI PORRAAA OLHA SEU RESULTADO: {quadrado(n)}")
```

```python
resultados = map(quadrado, numeros)
list(resultados)
```

```python
def celsius_para_fahrenheit(c):
    return (c * 9/5  ) + 32
```

```python
temperaturas_celsius = [0, 20, 30, 100]
```

```python
temperaturas_fahrenheit = map(celsius_para_fahrenheit, temperaturas_celsius)
list(temperaturas_fahrenheit)
```

```python
def normalizar_nome(nome):
    return nome.strip().lower()  # o .strip tira os espacos antes da palavra e depois dela, mas nao os espacos do meio, se quiser tirar os espacos do meio use um replace
                                 # ex: ???.replace(" ". "")  
```

```python
nomes = ["  Skibidi", " TOILET", "GYAAATTTTT", "eu sou muito bom em python"]
```

```python
nomes_normais = map(normalizar_nome, nomes)
print(list(nomes_normais))
```
