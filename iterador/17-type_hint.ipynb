```python
class Estudante:
    def __init__(self, nome: str, notas: list[float]) -> None: # cada estudante tem isso, ou seja, uma lista de notas em float e um nome em string
        self.nome = nome    # essa coisa aqui em     cima siginifica o retorno, ou seja se eu quero definir que o retorno da funcao deve ser um int voce escreve, antes dos :  "  -> int   " 
        self.notas = notas

    def media(self) -> float:
        return sum(self.notas) / len(self.notas)
    
```

```python
def alunos_acima_da_media(alunos: list[Estudante],media_minima: float,) -> list[str]:
    return [aluno.nome for aluno in alunos if aluno.media() > media_minima]
```

```python
alunos = [

    Estudante("Alice", [8.0, 7.8, 5.0, 6.6]),
    Estudante("Alianca ku", [5.0, 7.8, 4.0, 3.6]),
    Estudante("XUXA meneguel", [10.0, 10.0, 10.0, 10.0]),
    Estudante("Thaisa", [10.0, 8.8, 9.0, 8.6]),

]
```

```python
resultados = alunos_acima_da_media(alunos, 7.5)
resultados
```
