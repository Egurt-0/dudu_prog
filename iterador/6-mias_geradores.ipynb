```python
# gerador de expressao
quadrados = (x ** 2 for x in range(19))
for quadrado in quadrados:
    print(quadrado)
```

```python
#  iterador infinito
def contador_infinito():
    i = 0
    while True:
        yield i 
        i += 1

contador = contador_infinito()
for _ in range(19):
    print(next(contador))
```

```python
# coleta de valores usando send()
def gerado_soma():
    total = 0 
    while True:
        valor = yield total # recebe o valor e retorna o total acumulado
        if valor is not None:
            total += valor

soma = gerado_soma()
next(soma) # inicializa o gerador
print(soma.send(10))
print(soma.send(5))
print(soma.send(7))
```

```python
# tratamento de execoes
def gerado_execoes():
    try:
        while True:
            valor = yield
            print(f"Valor recebido: {valor}")
    except ValueError:
        print("Valor invalido meu brother, nao passa aqui nao camarada")

g = gerado_execoes()
next(g)
g.send(10)
g.throw(ValueError)
```

```python
# cascata de geradores
def mutiplica_por_dois(iterable):
    for i in iterable:
        yield i * 2

def adicionar_cinco(iterable):
    for i in iterable:
        yield i + 5

numeros = range(5) # 0, 1, 2 ,3 ,4
resultado = adicionar_cinco(mutiplica_por_dois(numeros))
for r in resultado:
    print(r)
```
