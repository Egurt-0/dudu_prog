```python
def linhas_numero_palavras(arquivo, numero,):
    with open(arquivo, 'r', encoding='utf-8') as file:
        for linha in file:
            if len(linha.split()) == numero:
                yield linha.strip()
```

```python
arquivo = 'texto.txt'
numero_palavras = 3
for linha in linhas_numero_palavras(arquivo, numero_palavras):
    print(linha)
```
