```python
import re
```

```python
def extrair_urls(texto):
    padrao_url = r'http?://[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}'
    urls = re.findall(padrao_url, texto)
    return urls
```

```python
texto_exemplos = """
    Visite nosso site em http://www.dudu.com e
    http://blog.exemplo.org para mais informacoes do produto.
    Entre em contato conosco em https://contato.skbidi.com.br
"""



urls_extraidas = extrair_urls(texto_exemplos)
print(urls_extraidas)
```

```python
def validar_senha(senha):
    padrao_senha = r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*?&])[A-Za-z\d@$!%*?&]{8,}$'
    return bool(re.match(padrao_senha, senha))
```

```python
senhas_teste = ["senhaFraca", "Senha123", "Senha@123", "senha@123", "SENHA@123", "Senha123!"]
for senha in senhas_teste:
    print(f"Senha: {senha}, Valida: {validar_senha(senha)}")
    
```

```python
def contar_palavras(texto):
    padrao_palavras = r'\b\w+\b'
    palavaras  = re.findall(padrao_palavras, texto)
    return len(palavaras)
```

```python
texto_exemplo = "ol a undo es sa porraa quiumex emplo"
numero_palavras = contar_palavras(texto_exemplo)
print(numero_palavras)
```
