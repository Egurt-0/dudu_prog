Yield
- palavra chave para criar geradores.quando voce usa ele em uma funcao, voce transforma a funcao em um gerador
que e um tipo especial de iterador.
- retrun: quando e usado em uma funcao e encerrada e o valor e retornado
- Yield: pode ser chamada novamente e retomar a execucao do ponto aonde foi pausada, permitindo a producao de uma sequencia de valores ao longo do tempo

```python
def numero_impares(n):
    for i in range(1, n+1, 2):
        yield i
```

```python
for numero in numero_impares(10):
    print(numero)
```
