zip
- combinar multiplos iteraveis, sendo lista ou tupla, em um unico iteravel

```python
nomes = ["Alice", "Bobinho", "Charlie", "guimaraes"]
idades = [25, 30 ,48, 67]
```

```python
for nome, idade in zip(nomes, idades):
    print(f"Nome: {nome}, Idade: {idade} ")
```

```python
pessoas  = dict(zip(nomes, idades))
pessoas
```

```python
pessoas_30_acima = {nome: idade for nome, idade in pessoas.items() if idade > 30}
pessoas_30_acima
```
