Async é um tipo de programação que pode ser executado de na bloquante, facilitando a execução de multiplas operações simultaneas


```python
import asyncio
import aiohttp
import pprint
```

```python
async def tarefa(nome, tempo):
    print(f'Tarefa {nome} foi iniciada')
    await asyncio.sleep(tempo)
    print(f"Tarefa {nome} foi concluida")
```

```python
async def main():
    await asyncio.gather(
        tarefa("A", 7),
        tarefa("B", 22) # aqui estou criando exemplos de tarefas, ali encima tem uma funcao "tarefa", tem nome e tempo, qui estou apenas criandop tarefas
        # e definindo o nome e o tempo delas, pois e obrigatorio preencher todas as caracteristicas ja colocadas.
    )
```

```python
try:
    asyncio.get_running_loop()
except RuntimeError:
    asyncio.run(main())
```

```python
asyncio.ensure_future(main())


# ignore toda essa bagaca para fazer ele rodar, oque e importante e o output
# o output comeca com o print das duas tarefas pois sua funcao pede que quando ela comecar a rodar, ela de o print do nome da tarefa e que ela foi iniciada
# mas o foda e a segunda parte, quando voce consegue ver que o segundo print, o print da tarefa ser concluida teve um intervalo de tempo, o intervalo que voce definiu quando criou a tarefa
# esse asyncio.sleep(tempo) faz com que ele espere um "tempo", que nesse caso voce definiu quando criou a tarefa, para dai ele executar oque voce pediu, nesse caso um print("tarefa {nome} foi concluida")
```

```python
async def buscar_usuario(session, usuario_id):
    async with session.get(f'https://jsonplaceholder.typicode.com/users/{usuario_id}') as resposta: # esse link e uma mini bibliotecas com alguns exemplos de ids, bom para fazer esses testexinhos
        return await resposta.json()
```

```python
async def main():
    async with aiohttp.ClientSession() as session:
        usuario_ids = range(1, 6)
        tarefas = [buscar_usuario(session, uid) for uid in usuario_ids]

        # guarda a conclusao de todas as tarefas
        usuarios = await asyncio.gather(*tarefas)
        for u in usuarios:
            print(f"ID: {u['id']} | Nome: {u['name']} | Email: {u['email']}") # formatacao que vi no chatGPT mas e bem d boa de enteder
```
