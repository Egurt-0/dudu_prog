```python
class VendasIterator:
    def __init__(self, arquivo): 
        self.file = open(arquivo, 'r', encoding='utf-8')

    def __iter__(self):
        return self

    def __next__(self):
        linha = self.file.readline()
        if not linha:
            self.file.close()
            raise StopIteration
        
        linha = linha.strip()
        if not linha:  # pula linha vazia
            return self.__next__()

        partes = linha.split(',')
        if len(partes) != 2:  # se não tiver categoria e valor
            print(f"Linha ignorada (formato inválido): {linha}")
            return self.__next__()

        categoria, valor = partes
        return categoria, float(valor)

                                                                                               

    
```

```python
def gerador_de_somar_vendas(arquivo):
    total_vendas = 0
    for categoria, valor in VendasIterator(arquivo):
        total_vendas += valor
        yield categoria, valor, total_vendas
```

```python
contagem_de_vendas = {}
arquivo = "vendas.txt"
print("processando vendas...\n")

for categoria, valor, total in gerador_de_somar_vendas(arquivo):
    print(f"Categoria: {categoria}, Valor R$: {valor:.2f}, Total: {total:.2f}")
    if categoria not in contagem_de_vendas:
        contagem_de_vendas[categoria] = 0
    contagem_de_vendas[categoria] += 1

print("\nResumo por categoria:")
for categoria, contagem in contagem_de_vendas.items():
    print(f"Categoria: {categoria} → {contagem} vendas")

```
