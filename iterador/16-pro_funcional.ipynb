```python
# funcao de alta ordem - uma funcao que recebe outra como argumento
def quadrado(x):
    return x*x


def aplicar_funcoes(funcao, lista): # essa funcao recebe outra funcao e uma lista
    return [funcao(x) for x in lista]
```

```python
lista_nuemeros = [1, 2, 3, 4, 5, 6, 7852]
resultado = aplicar_funcoes(quadrado, lista_nuemeros) # aqui nos definimos qual e a funcao que ela recebe e qual a lista
resultado
```

```python
#dados exemplos
usuarios = [
    {"nome": "Liz Macedo", "idade": 17},
    {"nome": "LUIZ AMARAL", "idade": 68},
    {"nome": "Liz Muller", "idade": 16},
]
```

```python
def exatrair_dados(usuarios):
    return list(map(lambda u: u["nome"], usuarios))


def filtrar_maior_de_idade(usuarios, idade):
    return list(filter(lambda u: u["idade"] >= idade, usuarios))

def media_idades(usuarios):
    if not usuarios:
        return 0
    total_idade = sum(map(lambda u: u["idade"], usuarios))
    return total_idade / len(usuarios)

```

```python
usuarios_maiores_de_idade = filtrar_maior_de_idade(usuarios, 17)
media_idades(usuarios)
```

```python
from functools import lru_cache
```

```python
@lru_cache(maxsize=None)
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci( n - 2)
```

```python
for i in range(10):
    print(f"fibonacci{i} = {fibonacci(i)}")
```
