MetaClasse
- cria uma classe de forma dinamica.
- pode ser util na criacao de frameworks aonde as classes precisam ter comportamentos especificos

```python
class MeuMeta(type):
    def __new__(cls, nome, bases, dct):
        dct['novo_atributo'] = 'Valor adicionado pela meta classe'
        return super().__new__(cls, nome, bases, dct)
        
```

```python
class MinhaClasse(metaclass=MeuMeta):
    pass
```

```python
obj = MinhaClasse()
obj.novo_atributo
```

```python
class Validador_dados(type):# toa classe que tiver (type) vai ser uma metaclasse
    def __new__(cls, nome, bases, dtc):
        validacoes = dtc.get( 'validacoes ', {})

        for attr, tipo in validacoes.items():
            if not callable(tipo):
                raise TypeError(f"Deu erro parceiro, o tipo de validacao para {attr} deve ser uma funcao")
            
            # funcao de validacao
            def valida_func(self, value, attr=attr, tipo=tipo):
                if not isinstance(value, tipo):
                    raise ValueError(f"{attr} deve ser do tipo {tipo.__name__}")
                self.__dict__[attr] = value
                
            # renomeia a funcao para evitar problemas de escopo
            valida_func.__name__ = f"set_{attr}"
            dtc[f"set_{attr}"] = valida_func

        return  super().__new__(cls, nome, bases, dtc) # super.__new__ isso serve para realmente criar a metaclasse, voce so passa super.__new__(todas as coisas que voce colocou ali em cima) 
```

```python
class Usuario(metaclass=Validador_dados):
    valiacoes = {
        'nome': str,
        'idade': int
    }

    def __init__(self, nome, idade):
        self.set_nome = nome
        self.set_idade = idade
        
```

```python
try:
    user = Usuario("Alice", 30)
    print(f"nome: {user.set_nome}, idade: {user.set_idade}") # nesse caso esse try e except era para quando eu usasse o segundo user.set_idade eu iria colocar "30" e entao ele daira o erro de ValueError pois idade = int sendo que defini idade como uma str
    user.set_idade = 30 # mas eu preferi deixar correto
except ValueError as e:
    print(e)
    
```
